```python
import sqlite3
import csv

def create_table(cursor):
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS cities (
            city_name TEXT,
            latitude REAL,
            longitude REAL
        )
    ''')

def insert_record(cursor, record):
    cursor.execute('INSERT INTO cities (city_name, latitude, longitude) VALUES (?, ?, ?)', record)

def load_records_from_csv(filename):
    records = []
    with open(filename, 'r') as csvfile:
        csvreader = csv.reader(csvfile)
        next(csvreader)  # Skip the header row
        for row in csvreader:
            city_name = row[0] + ' ' + row[2]
            latitude = float(row[6])
            longitude = float(row[7])
            records.append((city_name, latitude, longitude))
    return records

def main():
    conn = sqlite3.connect('cities-all.db')
    cursor = conn.cursor()

    create_table(cursor)

    csv_filename = 'uscities.csv'
    records_to_insert = load_records_from_csv(csv_filename)

    for record in records_to_insert:
        insert_record(cursor, record)

    conn.commit()
    conn.close()
    print("Records inserted successfully.")


if __name__ == "__main__":
    main()
```
