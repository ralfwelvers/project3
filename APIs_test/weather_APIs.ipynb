```python
# Dependencies
import requests
import json
from api_keys import weather_api_key
import pandas as pd
from pprint import pprint
```

```python
# Open Weather Map
# Current Weather

# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"

units = "imperial"

# This is Chicago
lat = 41.8781
lon = -87.6298

# Build partial query URL
query_url = url + "appid=" + weather_api_key + "&units=imperial" + "&lat=" + str(lat) + "&lon=" + str(lon)

response = requests.get(query_url).json()

pprint(response)
```
